```python
import csv

def generate_insert_statements(csv_data: str):
    reader = csv.DictReader(csv_data.strip().splitlines())
    location_inserts = []
    id_counter = 1  # AUTO_INCREMENT를 흉내 내기 위해 사용 (실제 DB에선 생략)

    for row in reader:
        print(row)
        source_id = row["연번"]
        name = row["상호"]
        full_address = row["도로명주소"]
        lat = round(float(row["위도"]), 2)
        lng = round(float(row["경도"]), 2)

        # 주소에서 시(city)와 구(district) 분리
        parts = full_address.split()
        city = parts[0] if len(parts) > 0 else None
        district = parts[1] if len(parts) > 1 else None

        insert_sql = f"""INSERT INTO recycle_locations (address, city, district, latitude, longitude, name, source_id)
VALUES ('{full_address}', '{city}', '{district}', {lat}, {lng}, '{name}', '{source_id}');"""
        location_inserts.append(insert_sql)

        id_counter += 1  # 다음 레코드를 위한 ID 증가 (실제 INSERT에는 사용되지 않음)

    return "\n".join(location_inserts)
  
def generate_insert_medicine(csv_data: str):
    reader = csv.DictReader(csv_data.strip().splitlines())
    location_inserts = []
    id_counter = 1  # AUTO_INCREMENT를 흉내 내기 위해 사용 (실제 DB에선 생략)

    for row in reader:
        print(row)
        full_address = row["도로명 주소"]
        city = full_address.split()[0] if len(full_address.split()) > 0 else None
        district = full_address.split()[1] if len(full_address.split()) > 1 else None
        name = row["관리번호"]
        lat = round(float(row["위도"]), 2)
        lng = round(float(row["경도"]), 2)
        source_id = row["연번"]

        insert_sql = f"""INSERT INTO recycle_locations (address, city, district, latitude, longitude, name, source_id)
VALUES ('{full_address}', '{city}', '{district}', {lat}, {lng}, '{name}', '{source_id}');"""
        location_inserts.append(insert_sql)

        id_counter += 1  # 다음 레코드를 위한 ID 증가 (실제 INSERT에는 사용되지 않음)

    return "\n".join(location_inserts)



# 예시 CSV 데이터

# with open('yeonsu_medicine.csv', mode='r', encoding='utf-8') as f:
#   csv_data = f.read()
#   # SQL 생성
#   sql_output = generate_insert_statements(csv_data)
#   print(sql_output)
  
with open('yeonsu_clothing.csv', mode='r', encoding='utf-8') as f:
  csv_data = f.read()
  print(generate_insert_medicine(csv_data))

```
